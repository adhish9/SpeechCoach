```python
def click_save(_):
    if name.value == '':
        print('Please enter name first!')
    else:
        global name_saved
        name_saved = name.value
        
        print("Click Record and read the following:\n Please call Stella."
              "Ask her to bring these things with her from the store: "
              "Six spoons of \nfresh snow peas, five thick slabs of blue cheese, "
              "and maybe a snack for her brother Bob. \n"
              "We also need a small plastic snake and a big toy frog for the kids. \n" 
              "She can scoop these things into three red bags," 
              "and we will go meet her Wednesday at the train station.")
        
#         if os.path.isdir("record/"+name.value) == True:
#             print("Hello again! "+name.value+',')
#         else:
#             print("Hello! "+name.value+',') 
#             try:
#                 os.mkdir("record/"+name.value)
#             except OSError:
#                 print ("Creation of the directory %s failed" % path)
                       
        
def click_record(_):
    if name_saved == '':
        print('Save name first!') 
    else:
        global filename
        filename = str(time.time())
        print("Listening.....")
        with m as source:
            audio = r.listen(source)
        try:
            # for testing purposes, we're just using the default API key
            # to use another API key, use `r.recognize_google(audio, key="GOOGLE_SPEECH_RECOGNITION_API_KEY")`
            # instead of `r.recognize_google(audio)`
            txt = r.recognize_google(audio, None, "en-IN")
            print("You said:\n"+txt)
            file = open("record/"+name_saved+filename+"_googlesr.txt","w") 
            file.write(txt) 
        except sr.UnknownValueError:
            print("Could not understand audio, Retry")
        except sr.RequestError as e:
            print("Could not request results from Google Speech Recognition service; {0}".format(e))

        with open("record/"+name_saved+filename+"_microphone.wav", "wb") as f:
            f.write(audio.get_wav_data())
            
#         print("Hello! "+name.value+',') 
#         print("Read the following:\n Please call Stella. 
#         Ask her to bring these things with her from the store: 
#         Six spoons of \nfresh snow peas, five thick slabs of blue cheese, 
#         and maybe a snack for her brother Bob. We also need a small plastic 
#         snake and a big toy frog for the kids. She can scoop these things into three red bags, 
#         and we will go meet her Wednesday at the train station.")
            
def click_retry(_):
    HTML("<script>Jupyter.notebook.kernel.restart()</script>")
    
    clear_output()
    #VBox([HBox([name,save]), first_line, second_line])
    show_ui()
    click_save(_)
    #click_record(_)

def click_stop(_):
    HTML("<script>Jupyter.notebook.kernel.restart()</script>")
    clear_output()
    #VBox([HBox([name,save]), first_line, second_line])
    show_ui()
    print("Stopped recording. Try again.")


def click_score(_):
    mysp=__import__("my-voice-analysis")
    print("---------Score----------")
    #import myspsolution as mysp
    if os.path.exists("record/"+name_saved+filename+"_microphone.wav") == True:
        p=name_saved+filename+"_microphone" # Audio File title  (.wav)
        c=r"/home/madtitan/Documents/Adhish/Speech/record/" # Path to the Audio_File directory (Python 3.7)
        mysp.mysptotal(p,c)

#     c=r"/home/madtitan/Documents/Adhish/Speech/record/"
#     p="trial11564629432.756363_microphone"
#     mysp.mysptotal(p,c)    
#     p="hindi1" #Audio File title
#     p="sushantmicrophone-results"
#     p="salonimicrophone-results"
#     mysp.myspgend(p,c)
#     c=r"/home/madtitan/Documents/Adhish/Speech/audio/hindi"
         
    else:
        print("Please record yourself first.")
```

main/init

```python
r = sr.Recognizer()
m = sr.Microphone()

#Adjust mic
with m as source:
    r.adjust_for_ambient_noise(source)

#display(Image("/home/madtitan/Documents/Adhish/Speech/icons/microphone.png"))


def show_ui():
    #print(1)
    name = widgets.Text(
           value='',
           description='Name',)

    
    save = Button(description="Save")
    record = Button(description='Record') 
    stop = Button(description='Stop')
    score = Button(description='Score')
    retry = Button(description='Retry')


    save.on_click(click_save)
    record.on_click(click_record)
    retry.on_click(click_retry)
    stop.on_click(click_stop)
    score.on_click(click_score)


    first_line = HBox([record, stop])
    second_line = HBox([score, retry])

    print('\n\n    Speech \n\n')
    display.display(VBox([HBox([name,save]), first_line, second_line]))

    
name_saved = ''
filename = ''

show_ui()
```
